**Rumus:**

Untuk $n+1$ titik data, polinom Hermite derajat $2n+1$ adalah:

$$H_{2n+1}(x) = \sum_{j=0}^{n} \bigl[f(x_j)\, h_j(x) + f'(x_j)\, \tilde{h}_j(x)\bigr]$$

dengan basis Lagrange $L_j(x) = \displaystyle\prod_{k \neq j} \dfrac{x - x_k}{x_j - x_k}$, dan:

$$h_j(x) = \bigl[1 - 2(x - x_j)L'_j(x_j)\bigr]\, L_j(x)^2, \qquad \tilde{h}_j(x) = (x - x_j)\, L_j(x)^2$$

Semua komputasi dilakukan secara **simbolik** menggunakan `sympy`.
