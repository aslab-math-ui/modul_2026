## Soal 2 — Polinom Hermite via Metode Lagrange &nbsp; *(30 poin)*

Gunakan **Metode Interpolasi Hermite melalui Polinom Lagrange** untuk menemukan polinom Hermite $H_{2n+1}(x)$ yang cocok dengan titik-titik pada soal 1. Tampilkan hasil polinomnya menggunakan `sympy.pprint`.

**Rumus:**

Untuk $n+1$ titik data, polinom Hermite derajat $2n+1$ adalah:

$$H_{2n+1}(x) = \sum_{j=0}^{n} \bigl[f(x_j)\, h_j(x) + f'(x_j)\, \tilde{h}_j(x)\bigr]$$

dengan basis Lagrange $L_j(x) = \displaystyle\prod_{k \neq j} \dfrac{x - x_k}{x_j - x_k}$, dan:

$$h_j(x) = \bigl[1 - 2(x - x_j)L'_j(x_j)\bigr]\, L_j(x)^2, \qquad \tilde{h}_j(x) = (x - x_j)\, L_j(x)^2$$

Semua komputasi dilakukan secara **simbolik** menggunakan `sympy`.

## Soal 3 — Polinom Hermite via Hermite Divided Difference &nbsp; *(30 poin)*

Gunakan **Metode Hermite Divided Difference** untuk menemukan polinom Hermite.  
Tampilkan **tabel divided difference** dan **hasil polinomnya** menggunakan `sympy.pprint`.

**Algoritma:**

1. Buat array $z$ dengan setiap titik $x_j$ dimunculkan **dua kali** (*doubled nodes*):  
   $z = [x_0, x_0,\, x_1, x_1,\, \ldots,\, x_n, x_n]$

2. Inisialisasi: $f[z_i] = f(x_j)$

3. Divided difference orde pertama untuk titik bersamaan ($z_i = z_{i-1}$): gunakan $f'(x_j)$

4. Untuk orde lebih tinggi, gunakan rumus standar:  
   $$f[z_{i-k},\ldots,z_i] = \dfrac{f[z_{i-k+1},\ldots,z_i] - f[z_{i-k},\ldots,z_{i-1}]}{z_i - z_{i-k}}$$

5. Bangun polinom dari koefisien diagonal (bentuk Newton):  
   $$H(x) = f[z_0] + f[z_0,z_1](x-z_0) + f[z_0,z_1,z_2](x-z_0)(x-z_1) + \cdots$$

## Soal 6 — Metode Neville &nbsp; *(10 poin)*

Menggunakan **Metode Neville** (hanya menggunakan data $f(x)$, **tanpa** data $f'(x)$), cari nilai aproksimasi fungsi pada titik $x = 0{,}13$. Tampilkan tabel Neville lengkapnya.

**Rumus rekursi Neville:**

Inisialisasi: $Q_{i,0} = f(x_i)$

$$Q_{i,j} = \frac{(x - x_{i-j})\, Q_{i,j-1} - (x - x_i)\, Q_{i-1,j-1}}{x_i - x_{i-j}}$$

Hasil aproksimasi: $Q_{n,n}$ (elemen pojok kanan bawah tabel).
